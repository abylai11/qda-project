Confidence intervals

Assuming all images we have are not defective -> create CI of the covariates
- Done before addressing 'outliers'
- Might be good (defective will be outside CI) or useless (too big interval because of outliers)

```python
# on numeric vars
dataNum = df.select_dtypes(include=np.number)
dataNum=dataNum.drop(columns=['ID','Part_ID'])
dataNum.head(10)
```

```python
# Normality assumption
for col in dataNum.columns.tolist():
  # Shapiro-Wilk test
  _, p_value_SW = stats.shapiro(dataNum[col])
  print('p-value of the Shapiro-Wilk ', col, ' test: %.3f' % p_value_SW)
```
