```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import qda
```

```python
# Define a range of values for beta
delta = np.linspace(0, 4, 100)
ARL=200
alpha=1/ARL
k=stats.norm.ppf(1-alpha/2)
n=[5,10]

# Compute the corresponding beta values
beta1 = stats.norm.cdf(k - delta*np.sqrt(n[0])) - stats.norm.cdf(-k - delta*np.sqrt(n[0]))
beta2 = stats.norm.cdf(k - delta*np.sqrt(n[1])) - stats.norm.cdf(-k - delta*np.sqrt(n[1]))
# Plot the beta values
plt.plot(delta, beta1)
plt.plot(delta,beta2)
plt.xlabel('Delta')
plt.ylabel('Beta')
plt.title('Operating characteristic curve')
plt.show()
```

```python
ARL1=1/(1-beta1)
ARL2=1/(1-beta2)
h=3
ATS1=ARL1*h
ATS2=ARL2*h
# Plot the beta values
plt.plot(delta, ATS1)
plt.plot(delta,ATS2)
plt.xlabel('Delta')
plt.ylabel('ATS')
plt.title('ATS curve')
plt.show()
```

```python
# Define a range of values for beta
n= np.linspace(2, 50, 100)
ARL=200
alpha=1/ARL
k=stats.norm.ppf(1-alpha/2)
delta=[1,2]

# Compute the corresponding beta values
beta1 = stats.norm.cdf(k - delta[0]*np.sqrt(n)) - stats.norm.cdf(-k - delta[0]*np.sqrt(n))
beta2 = stats.norm.cdf(k - delta[1]*np.sqrt(n)) - stats.norm.cdf(-k - delta[1]*np.sqrt(n))
# Plot the beta values
plt.plot(n, beta1)
plt.plot(n,beta2)
plt.xlabel('n')
plt.ylabel('Beta')
plt.title('Operating characteristic curve')
plt.show()
```

```python
ARL1=1/(1-beta1)
ARL2=1/(1-beta2)
h=3
ATS1=ARL1*h
ATS2=ARL2*h
# Plot the beta values
plt.plot(n, ATS1)
plt.plot(n,ATS2)
plt.xlabel('n')
plt.ylabel('ATS')
plt.title('ATS curve')
plt.show()
```
