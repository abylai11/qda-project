```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from scipy import stats
import qda
```

```python
# Define a range of values for beta
delta = np.linspace(0, 2, 100)
ARL=250
alpha=1/ARL
k=stats.norm.ppf(1-alpha/2)
n=[4,8]

# Compute the corresponding beta values
beta1 = stats.norm.cdf(k - delta*np.sqrt(n[0])) - stats.norm.cdf(-k - delta*np.sqrt(n[0]))
beta2 = stats.norm.cdf(k - delta*np.sqrt(n[1])) - stats.norm.cdf(-k - delta*np.sqrt(n[1]))
# Plot the beta values
plt.plot(delta, beta1)
plt.plot(delta,beta2)
plt.xlabel('Delta')
plt.ylabel('Beta')
plt.title('Operating characteristic curve')
plt.show()
```

```python
ARL1=1/(1-beta1)
ARL2=1/(1-beta2)

# Plot the beta values
plt.plot(delta, ARL1)
plt.plot(delta,ARL2)
plt.xlabel('Delta')
plt.ylabel('ARL')
plt.title('ARL curve')
plt.show()
```

```python
ARL1=1/(1-beta1)
ARL2=1/(1-beta2)

# Plot the beta values
plt.plot(n, ARL1)
plt.plot(n,ARL2)
plt.xlabel('n')
plt.ylabel('ARL')
plt.title('ARL curve')
plt.show()
```
